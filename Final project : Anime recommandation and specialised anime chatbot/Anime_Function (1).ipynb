```python
# keep using anime23
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

def find_similar_byanime(anime_name, anime_df, top_n=10):
    # Filter the anime DataFrame to get the data for the given anime
    anime_data = anime_df[~anime_df['type'].isin(['movie', 'Movie'])]
    anime_data = anime_df[anime_df["title"] == anime_name]


    if anime_data.empty:
        return "Anime not found in the anime_df."

    # Create a user-anime matrix based on the provided anime data
    user_anime_matrix = pd.pivot_table(data=anime_df,
                                       values='my_score',
                                       index='user_id',
                                       columns='anime_id',
                                       fill_value=0)

    # Compute the pairwise cosine similarity between the given anime and all other anime
    anime_vector = user_anime_matrix.loc[:, anime_data["anime_id"].values].values
    similarities = cosine_similarity(anime_vector.T, anime_vector.T)

    # Get the indices of the top N most similar anime (excluding the given anime itself)
    similar_anime_indices = similarities.argsort()[0][-2:-top_n-2:-1]

    # Get the anime IDs of the top N most similar anime
    similar_anime_ids = user_anime_matrix.columns[similar_anime_indices]

    # Get the titles of the top N most similar anime using the anime_df
    similar_anime_names = [anime_df.loc[anime_id, "title"] for anime_id in similar_anime_ids]

    return similar_anime_names

```

```python
import pandas as pd
from sklearn.metrics.pairwise import pairwise_distances

def collaborative_filtering(user_name, anime_df, top_n=10):
    # Pivot the dataframe to create user-item matrix
    user_item_matrix = anime_df.pivot(index='username', columns='anime_id', values='my_score').fillna(0)

    # Calculate user-user similarity
    user_similarity = pairwise_distances(user_item_matrix, metric='correlation')

    # Find the index of the target user
    user_index = user_item_matrix.index.get_loc(user_name)

    # Find similar users to the target user
    similar_users_indices = user_similarity[user_index].argsort()[1:top_n+1]

    # Get the ratings of similar users
    similar_users_ratings = user_item_matrix.iloc[similar_users_indices]

    # Calculate the average rating of each item across similar users
    item_ratings = similar_users_ratings.mean()

    # Filter out items already rated by the target user
    user_ratings = user_item_matrix.loc[user_name]
    unseen_items = user_ratings[user_ratings == 0].index

    # Sort items by average rating and recommend the top N
    recommendations = item_ratings.loc[unseen_items].sort_values(ascending=False).head(top_n)

    return recommendations

# Example usage
user_name = "MistButterfly"
top_anime_recommendations = collaborative_filtering(user_name, anime23, top_n=10)
print("Top anime recommendations:")
print(top_anime_recommendations)
```

```python
import pandas as pd
from sklearn.metrics.pairwise import pairwise_distances

def collaborative_filtering(user_name, anime_df, top_n=10):
    # Pivot the dataframe to create user-item matrix
    user_item_matrix = anime_df.pivot(index='username', columns='anime_id', values='my_score').fillna(0)

    # Calculate user-user similarity
    user_similarity = pairwise_distances(user_item_matrix, metric='correlation')

    # Find the index of the target user
    user_index = user_item_matrix.index.get_loc(user_name)

    # Find similar users to the target user
    similar_users_indices = user_similarity[user_index].argsort()[1:top_n+1]

    # Get the ratings of similar users
    similar_users_ratings = user_item_matrix.iloc[similar_users_indices]

    # Calculate the average rating of each item across similar users
    item_ratings = similar_users_ratings.mean()

    # Filter out items already rated by the target user
    user_ratings = user_item_matrix.loc[user_name]
    unseen_items = user_ratings[user_ratings == 0].index

    # Sort items by average rating and recommend the top N
    recommendations = item_ratings.loc[unseen_items].sort_values(ascending=False).head(top_n)

    # Get the titles of recommended anime
    recommended_anime_ids = recommendations.index
    recommended_anime_titles = anime_df.loc[anime_df['anime_id'].isin(recommended_anime_ids) & (anime_df['username'] != user_name), 'title'].tolist()

    return recommended_anime_titles

# Example usage
user_name = "MistButterfly"
top_anime_recommendations = collaborative_filtering(user_name, anime23, top_n=10)
print("Top anime recommendations:")
print(top_anime_recommendations)
```

# TF-IDF Vectoriyer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

def get_top_similar_anime_by_synopsis(anime_df, anime_name, num):
    # Define a TF-IDF Vectorizer
    tfidf = TfidfVectorizer(stop_words='english')

    # Replace NaN values with an empty string
    anime_df["synopsis"] = anime_df["synopsis"].fillna("")

    # Construct the TF-IDF matrix
    tfidf_matrix = tfidf.fit_transform(anime_df["synopsis"])

    # Compute the cosine similarity matrix
    cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

    # Get the index of the anime matching the title
    idx = anime_df[anime_df['title'] == anime_name].index[0]

    # Get the pairwise similarity scores of all anime with that anime
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the anime based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the top similar anime indices
    sim_indices = [i[0] for i in sim_scores[1:num+1]]  # Exclude the first item (itself) and get top num similar

    # Return the top similar anime
    return anime_df.iloc[sim_indices][['title', 'img_url']]


```

```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display


def anime_info_dropdown(anime_df):
    # Create a dropdown list of anime titles
    dropdown_options = anime_df['title'].tolist()
    dropdown = widgets.Dropdown(options=dropdown_options, description='Select Anime:')

    # Define a function to display anime information when an anime is selected
    def display_info(selected_anime):
        anime_info = anime_df[anime_df['title'] == selected_anime]
        print(f"TITLE: {anime_info['title'].iloc[0]}")
        print(f"GENRE: {anime_info['genre'].iloc[0]}")
        print(f"EPISODES: {anime_info['episodes'].iloc[0]}")
        print(f"SCORE: {anime_info['score'].iloc[0]}")
        print(f"SYNOPSIS: {anime_info['synopsis'].iloc[0]}")

    # Display the dropdown and observe the selection
    display(dropdown)
    widgets.interact(display_info, selected_anime=dropdown)


```
