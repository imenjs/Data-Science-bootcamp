## 2.  The Game Code

```python
from random import randrange

easy_words = ["python","javaScript","java", "juby","swift", "php", "go", "kotlin",  "rust", "MmATLAB", "perl", "scala", "groovy"]

hard_words = ["germany", "brazil" , "japan", "canada" , "australia" , "mexico" ,"egypt", "sweden", "argentina", "tunisia"]

super_hard_words = ["jujutsukaisen" ,"attackontitan", "onepiece", "Bleach", "deathnote", "fullmetalalchemist", "dragonballz", "demonslayer", "hunterxhunter", "onepunchman",  "madeinabyss", "mobpsycho"]

# Display name of the game in Red
print("\033[1;31mWelcome to Hangman Game !!!!\033[0m")

# check dificulty and set varible accordently
Difficulty = int(input("please enter '1' for easy (coding Languages) , '2' for hard (country names), '3' for supper hard (anime) "))

if Difficulty == 1 :
    word = easy_words[randrange(len(easy_words))]
    mistake_count = 4
    mistake=4
    hangman = hangman4
elif    Difficulty == 2 :
    word = hard_words[randrange(len(hard_words))]
    mistake_count = 6
    mistake=6
    hangman = hangman6
elif    Difficulty == 3 :
    word = super_hard_words[randrange(len(super_hard_words))]
    mistake_count = 8
    mistake=8
    hangman = hangman8

# Initialize secret_word as a list of characters
secret_word = ["_" for _ in word]
print(f"The secret word is {' '.join(secret_word)}")
#print(f"The actual word is {word}")

# list of all letter the user will enter
list_letters = ""


while mistake_count > 0 and secret_word.count("_") != 0:
    letter = input("Please enter a letter: ")

    # check if player entred more than 1 letter or entred a number
    if len(letter) != 1 or not letter.isalpha():
        print("Please enter a valid single letter.")
        print("*************************************************")
        continue

    # Check if the letter has been guessed before
    if letter in list_letters:
        print("You already guessed that letter. Enter a new letter.")
        print("*************************************************")
        continue

    list_letters += letter

    found = False

    for index, i in enumerate(word):
        if i == letter:
            secret_word[index] = letter
            found = True
   # print(f"number of attempts left : {mistake_count} ")


    if not found:
        mistake_count -= 1
        print("oh no  !!!!!!!!!!!! ")
        print (hangman[mistake- mistake_count])
    print(f"number of attempts left : {mistake_count} ")

    print("Secret word :", ' '.join(secret_word))
    print("letter :", ' '.join(list_letters))
    print("*************************************************")

# Check if the player won or lost
if mistake_count > 0:
    print("Congratulations! You guessed the word:", word)
else:
    print("Sorry, you ran out of attempts. The correct word was:", word)
```
